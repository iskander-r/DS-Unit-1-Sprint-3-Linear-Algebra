## 1.1 Create a two-dimensional vector and plot it on a graph

```
# Imports
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
np_a =np.array([0.3, 0.5, 0.4])
np_b = np.array([4,5,5])
```

```
fig, ax = plt.subplots()
ax.grid()
plt.xlim(-0.2, 6)
plt.ylim(-0.2, 6)


for mycolor, myvector in [('crimson', np_b)]:
  plt.arrow(0,0,
            myvector[0],
            myvector[1],
            head_width=.02,
            head_length=.02,
            linewidth=3,
            color=mycolor)
```

## 1.2 Create a three-dimensional vecor and plot it on a graph

```
myvector = np.array([[0,0.1,0.1, .5, .5, .5]])
```

```
X, Y, Z, U, V, W = zip(*myvector)
```

```
# display that in 3D
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W)


ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.set_zlim([0, 1])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
